```python
data = pd.read_csv("data/id_kompetensi_flag_clean_stem.csv", sep=",")
data.isnull().sum()
```

```python
data = data.dropna()
data.isnull().sum()
```

```python
list_pegawai = list(set(data['ID_PEGAWAI']))
list_pegawai
```

```python
data.set_index('ID_PEGAWAI',inplace=True)
```

```python
list_pegawai = list(list_pegawai)
```

```python
list_pegawai[1]
```

```python
new_dict_kompotensi = {}
sep = '.'
for x in list_pegawai:
    temp = list(data.loc[x, 'KOMPETENSI'])    
    new_dict_kompotensi[x] = sep.join(temp)
df_kompetensi = {'ID_PEGAWAI': list(new_dict_kompotensi.keys()), 'KOMPETENSI': list(new_dict_kompotensi.values())}    
df = pd.DataFrame.from_dict(df_kompetensi)
df
```

```python
df.to_csv('data/grouping_pegawai.csv', index=False)
```
